```python
import pandas as pd
import numpy as np
```

```python
prices = pd.read_csv("data/sample_prices.csv")
prices
```

```python
returns = prices.pct_change()
returns
```

```python
returns = returns.dropna()
returns
```

```python
deviance = returns - returns.mean()
squared_deviance = deviance**2
squared_deviance
```

```python
np.sqrt(squared_deviance.mean())
```

```python
# monthly volatility
returns.std()
```

```python
num_of_obs = squared_deviance.shape[0]
np.sqrt(squared_deviance.sum()/(num_of_obs - 1))
```
